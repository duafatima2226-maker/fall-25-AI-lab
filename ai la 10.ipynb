```python
✅ ✔ Final Python Code 

# --- AI Lab 10 Task ---
# Data Preprocessing, Null Handling, Datatype Conversion

import pandas as pd
import numpy as np

# 1. Reading dataset (sample name dataset.csv)
data = pd.read_csv("dataset.csv")

print("Dataset Loaded Successfully!\n")

# 2. Print rows and columns
print("Rows:", data.shape[0])
print("Columns:", data.shape[1])

# 3. Check null values in each column
print("\nNull values in each column:")
print(data.isnull().sum())

# 4. Fill null values (numeric columns with mean, object columns with mode)
for col in data.columns:
    if data[col].dtype == 'object':        # if column type is string
        mode_val = data[col].mode()[0]
        data[col] = data[col].fillna(mode_val)
    else:
        mean_val = data[col].mean()
        data[col] = data[col].fillna(mean_val)

print("\nNull values after filling:")
print(data.isnull().sum())

# 5. Convert object columns to integer using factorize()
cat_columns = data.select_dtypes(include=['object']).columns

print("\nObject columns:", list(cat_columns))

data[cat_columns] = data[cat_columns].apply(lambda col: pd.factorize(col)[0])

# 6. Convert numeric columns to int64
for col in data.columns:
    data[col] = data[col].astype(np.int64)

print("\nUpdated Datatypes:")
print(data.dtypes)

# 7. Display first 5 rows
print("\nCleaned Data:")
print(data.head())
```
