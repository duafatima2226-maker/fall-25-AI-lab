```python
# --- AI Lab 9 Task Solution ---
# Data Loading & Data Exploration

import pandas as pd

# 1. Load dataset
data = pd.read_csv("students.csv")   # you can replace with any Kaggle dataset
print("Dataset Loaded Successfully!\n")

# 2. Print Top 5 rows
print("Top 5 Rows:\n")
print(data.head())

# 3. Print Bottom 5 rows
print("\nBottom 5 Rows:\n")
print(data.tail())

# 4. Print number of rows and columns
print("\nNumber of Rows:", data.shape[0])
print("Number of Columns:", data.shape[1])

# 5. Identify null values in dataset
print("\nNull Values in each Column:\n")
print(data.isnull().sum())

# 6. Fill null values using MODE for all columns
for col in data.columns:
    data[col] = data[col].fillna(data[col].mode()[0])

print("\nNull Values AFTER Filling:\n")
print(data.isnull().sum())

# 7. Print datatypes of each column
print("\nDatatypes of all Columns:\n")
print(data.dtypes)
```
